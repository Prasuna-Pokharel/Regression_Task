```python
#checking outliers in the dataset 
col_dict = {'Life_Expectancy':1,'Adult_Mortality':2,'Infant_Deaths':3,'HepatitisB':4,'Polio':5,'Tot_Exp':6,' HIV/AIDS':7,'Income_Comp_Of_Resources':8}

# Detect outliers in each variable using box plots.
plt.figure(figsize=(15,15))

for variable,i in col_dict.items():
                     plt.subplot(5,4,i)
                     plt.boxplot(df[variable],whis=1.5)
                     plt.title(variable)

plt.show()
```

* Anything that lies outside of lower and upper bound is an outlier. As expected, the boxplot confirms that there are a lot of outliers for all of these variables, including the goal variable, life expectancy, as can be shown visually.

```python
#printing out the number of outliers and its percentage

def outlier_count(col, data=df):
    print(15*'-' + col + 15*'-')
    q75, q25 = np.percentile(data[col], [75, 25])
    iqr = q75 - q25
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    print('Number of outliers: {}'.format(outlier_count))
    print('Percent of data that is outlier: {}%'.format(outlier_percent))
for col in col_dict:
    outlier_count(col)
```

```python
# treating the outliers 

features_with_outliers = [ 'Life_Expectancy','Adult_Mortality','Infant_Deaths','HepatitisB','Polio','Tot_Exp',' HIV/AIDS','Income_Comp_Of_Resources']
for feature in features_with_outliers:
    q1 = df[feature].quantile(0.25)#the median of the lower half 
    q3 = df[feature].quantile(0.75)#the median of the upper half 
    IQR = q3-q1#interquartile range
    lower_limit = q1 - (IQR*1.5)
    upper_limit = q3 + (IQR*1.5)
    df.loc[df[feature]<lower_limit,feature] = lower_limit
    df.loc[df[feature]>upper_limit,feature] = upper_limit
```

* There are 'object' type of features in the dataset. Lets explore them and try to convert them into numericals.

```python
# Encoding of Categorical Features
def encode_data(feature_name):

    ''' 

    This function takes feature name as a parameter and returns mapping dictionary to replace(or map) categorical 
    data with numerical data.

    '''

    mapping_dict = {}

    unique_values = list(df[feature_name].unique())

    for idx in range(len(unique_values)):

        mapping_dict[unique_values[idx]] = idx

    return mapping_dict

# categorical features in the dataset
df['Status'].replace({'Developing':0, 'Developed': 1}, inplace = True)
df['Country'].replace(encode_data('Country'), inplace = True)
```

# Performing cross validation with Randomized Search CV

```python
#importing randomized search CV
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 500, num = 5)]#The number of trees in the forest num=5 gap
max_features = ['auto', 'sqrt']#he number of features to consider when looking for the best split #auto sav lina khojcha tesle overfitting huna sakcha #total features ko square root
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]#The maximum depth of the tree
min_samples_split = [2, 5, 10, 15, 100]#The minimum number of samples required to split an internal node
min_samples_leaf = [1, 2, 5, 10]#the minimum number of samples required to be at a leaf node
#hyperparameters
#random forest ko parameter haru ho 
random_grid = {'n_estimators': n_estimators,#paramter lai variables assign gareko 
               
'max_features': max_features,
'max_depth': max_depth,
'min_samples_split': min_samples_split,
'min_samples_leaf': min_samples_leaf
}
random_rf = RandomizedSearchCV(RandomForestRegressor(), param_distributions=random_grid, cv=17, n_iter=10,n_jobs=-1)#cv= 17 divide equally 
random_rf.fit(X_train, y_train)
print(random_rf.best_estimator_)
print(random_rf.best_score_)
print(random_rf.best_params_)
```
